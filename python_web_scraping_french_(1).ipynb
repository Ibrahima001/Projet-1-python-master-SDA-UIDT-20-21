## Inspecter le code source HTML d'une page Web

![](https://i.imgur.com/mvBpQIP.png)

Comme mentionné précédemment, les pages Web sont écrites dans un langage appelé HTML (Hyper Text Markup Language). HTML est un langage assez simple composé de *tags* (également appelés *nodes* ou *elements*), par ex. `<a href="https://jovian.ai" target="_blank">Aller à Jovian</a>`. Une balise HTML comporte trois parties :

1. **Nom** : (`html`, `head`, `body`, `div`, etc.) Indique ce que représente la balise et comment un navigateur doit interpréter les informations qu'elle contient.
2. **Attributs** : (`href`, `target`, `class`, `id`, etc.) Propriétés de la balise utilisées par le navigateur pour personnaliser l'affichage d'une balise et décider de ce qui se passe lors des interactions de l'utilisateur.
3. **Enfants** : une balise peut contenir du texte ou d'autres balises ou les deux entre les segments d'ouverture et de fermeture, par exemple, `<div>Du contenu</div>`.

### Principaux référentiels pour un sujet

Revenons à notre énoncé initial du problème consistant à trouver les meilleurs référentiels pour un sujet donné. Avant d'analyser une page et de trouver les meilleurs référentiels, définissons une fonction d'assistance pour obtenir la page Web pour n'importe quel sujet.

> **QUESTION** : définissez une fonction `get_topic_page` qui télécharge la page Web GitHub pour un sujet donné et renvoie un objet BeautifulSoup représentant la page.

````python
def get_topic_page(topic):
    # Construct the URL
    topic_repos_url = 'https://github.com/topics/' + topic
    
    # Get the HTML page content using requests
    response = requests.get(topic_repos_url)
    
    # Ensure that the reponse is valid
    if response.status_code != 200:
        print('Status code:', response.status_code)
        raise Exception('Failed to fetch web page ' + topic_repos_url)
    
    # Construct a beautiful soup document
    doc = BeautifulSoup(response.text)
    
    return doc
````

````python
doc = get_topic_page('machine-learning')
````

````python
doc.title.text
````

````python
## Écriture d'informations dans des fichiers CSV

Créons une fonction d'assistance qui prend une liste de dictionnaires et les écrit dans un fichier CSV.

L'entrée de notre fonction sera une liste de dictionnaire de la forme :

```
[
   {'key1': 'abc', 'key2': 'def', 'key3': 'ghi'},
   {'key1': 'jkl', 'key2': 'mno', 'key3': 'pqr'},
   {'key1': 'stu', 'key2': 'vwx', 'key3': 'yza'}
   ...
]
```

La fonction va créer un fichier avec un nom donné contenant les données suivantes :

```
clé1,clé2,clé3
abc,def,ghi
jkl, mno, pqr
stu,vwx,yza

```
````

````python
def write_csv(items, path):
    # Open the file in write mode
    with open(path, 'w') as f:
        # Return if there's nothing to write
        if len(items) == 0:
            return
        
        # Write the headers in the first line
        headers = list(items[0].keys())
        f.write(','.join(headers) + '\n')
        
        # Write one item per line
        for item in items:
            values = []
            for header in headers:
                values.append(str(item.get(header, "")))
            f.write(','.join(values) + "\n")
````

> **QUESTION** : écrivez une fonction Python qui crée un fichier CSV (valeurs séparées par des virgules) contenant des détails sur les 25 principaux référentiels GitHub pour un sujet donné. Les principaux référentiels pour le sujet « machine-learning » se trouvent sur cette page : [https://github.com/topics/machine-learning](https://github.com/topics/machine-learning). Le fichier CSV de sortie doit contenir ces détails : nom du référentiel, nom d'utilisateur du propriétaire, no. d'étoiles, URL du référentiel.

````python
import requests
from bs4 import BeautifulSoup
base_url = 'https://gitub.com'

def scrape_topic_repositories(topic, path=None):
    """Get the top repositories for a topic and write them to a CSV file"""
    if path is None:
        path = topic + '.csv'
    topic_page_doc = get_topic_page(topic)
    topic_repositories = get_top_repositories(topic_page_doc)
    write_csv(topic_repositories, path)
    print('Top repositories for topic "{}" written to file "{}"'.format(topic, path))
    return path

def get_top_repositories(doc):
    """Parse the top repositories for a topic given a Beautiful Soup document"""
    article_tags = doc.find_all('article', class_='border rounded color-shadow-small color-bg-secondary my-4')
    topic_repos = [parse_repository(tag) for tag in article_tags]
    return topic_repos

def get_topic_page(topic):
    """Get the web page containing the top repositories for a topic as a Beautiful Soup document"""
    topic_repos_url = 'https://github.com/topics/' + topic
    response = requests.get(topic_repos_url)
    if response.status_code != 200:
        print('Status code:', response.status_code)
        raise Exception('Failed to fetch web page ' + topic_repos_url)
    return BeautifulSoup(response.text)    

def parse_repository(article_tag):
    """Parse information about a repository from an <article> tag"""
    a_tags = article_tag.h3.find_all('a')
    username = a_tags[0].text.strip()
    repo_name = a_tags[1].text.strip()
    repo_url = base_url + a_tags[1]['href'].strip()
    stars_tag = article_tag.find('a', class_='social-count float-none')
    star_count = parse_star_count(stars_tag.text.strip())
    return {'repository_name': repo_name, 'owner_username': username, 'stars': star_count, 'repository_url': repo_url}

def parse_star_count(stars_str):
    """Parse strings like 40.3k and get the no. of stars as a number"""
    stars_str = stars_str.strip()
    return int(float(stars_str[:-1]) * 1000) if stars_str[-1] == 'k' else int(stars_str)

def write_csv(items, path):
    """Write a list of dictionaries to a CSV file"""
    with open(path, 'w') as f:
        if len(items) == 0:
            return
        headers = list(items[0].keys())
        f.write(','.join(headers) + '\n')
        for item in items:
            values = []
            for header in headers:
                values.append(str(item.get(header, "")))
            f.write(','.join(values) + "\n")
````
